#DATA ANALYSIS AND VISUALIZATIONS


```python
cancelled_perc =df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('reservation_status count')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts())
```

#This bar chart provides an analysis of the reservation status count, specifically comparing the number of canceled and non-canceled reservations. Here's a detailed analysis based on the chart:

 Key Observations:
1. **Non-Canceled Reservations:**
   - The number of reservations that were not canceled is significantly higher than the number of canceled reservations.
   - The count is approximately 70,000 or more.

2. **Canceled Reservations:**
   - The count of canceled reservations is notably lower than non-canceled reservations.
   - The approximate count is around 40,000.

3. **General Trend:**
   - A substantial proportion of reservations were successfully completed or retained, with cancellations forming a smaller, but still significant, portion.

 Implications:
- **Customer Retention:**
  - The high count of non-canceled reservations indicates strong customer retention or successful booking completion.
  
- **Cancellation Rate:**
  - The cancellation count, while smaller, is still notable and could warrant investigation into reasons for cancellations (e.g., pricing, scheduling, external factors).

- **Operational Insights:**
  - If cancellations are preventable, addressing the causes could improve overall revenue and efficiency.


```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue ='is_canceled',data=df,palette='Blues')
legend_labels_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

#The chart compares reservation statuses across Resort and City Hotels. City Hotels have significantly more reservations overall, both canceled (dark bars) and non-canceled (light bars). Resort Hotels show fewer reservations, with cancellations forming a smaller proportion. City Hotels likely attract more customers but face higher cancellation rates.

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel= df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel =resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel =city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(8,4))
plt.title('Avergae Dail rate incity and resort hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=10)
plt.show()
```

#The line chart shows the average daily rates for Resort and City Hotels over time. Resort Hotels exhibit greater volatility in rates, with sharp spikes and fluctuations, especially between 2015 and 2017. City Hotels maintain relatively stable rates, with smaller variations. Resort Hotels generally have higher peaks in rates compared to City Hotels. Seasonal demand likely influences Resort Hotel pricing, while City Hotels appear to adopt more consistent pricing strategies.

```python
 df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month', hue ='is_canceled',data=df,palette='bright')
legend_labels_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',size=20)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation cancel',size=20)
plt.pie(top_10_country,autopct='%.2F',labels =top_10_country)
```
